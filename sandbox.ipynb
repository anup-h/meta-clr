```python
import cv2 as cv
import moviepy.editor as mp

clip = mp.VideoFileClip('youtube8m_rgb/ianXhsv5HQA.mp4')
duration = clip.reader.nframes
f = clip.get_frame(0)
prvs = cv.cvtColor(f, cv.COLOR_BGR2GRAY)
hsv = np.zeros_like(f)
hsv[...,1] = 255
result = []

for i in range(1, duration):
    f = clip.get_frame(i)
    if(i%5==0):
        curr = cv.cvtColor(f, cv.COLOR_BGR2GRAY)
        flow = cv.calcOpticalFlowFarneback(prvs,curr, None, 0.5, 3, 15, 3, 5, 1.2, 0)
        mag, ang = cv.cartToPolar(flow[...,0], flow[...,1])
        hsv[...,0] = ang*180/np.pi/2
        hsv[...,2] = cv.normalize(mag,None,0,255,cv.NORM_MINMAX)
        bgr = cv.cvtColor(hsv,cv.COLOR_HSV2BGR)
        result.append(bgr)
        prvs = curr

result = np.stack(result)
```

```python
f = open("durations.json","r")
durations = json.load(f)
clip_length = 50
buffer = clip_length
```

```python
clips = []
for k, v in durations.items():
    for i in range(v//(clip_length+buffer)):
        clips.append((k, i*(clip_length+buffer)))
```

```python
import random
random.seed(42)
random.shuffle(clips)
```
